```python

def custom_tokenizer(s):
    # Remove punctuation from the string using string.punctuation
    for char in s:
        if char in string.punctuation:
            s = s.replace(char, '')

    # Make the entire string lowercase
    s = s.lower()

    # Split the string at each space to make the list of tokens (uncleaned)
    tokens = s.split()

    stop_words = nltk.corpus.stopwords.words('english')

    # Save NLTK stop words to a variable
    stop_words = set(stopwords.words('english'))

    # Use list comprehension to create a list of tokens that are NOT stop words
    tokens_new = [token for token in tokens if token not in stop_words]


    # Create WordNetLemmatizer object
    wnl = nltk.stem.WordNetLemmatizer()

    # Lemmatize tokens
    lemmatized_tokens = [wnl.lemmatize(token) for token in tokens_new]

    return lemmatized_tokens

```

```python
stemmer = nltk.stem.PorterStemmer()
count_vectorizer = CountVectorizer(
    tokenizer = custom_tokenizer,
    stop_words='english',
    min_df=5,
    max_features=1000
)


```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score


# Create a MLPRegressor instance
nn_regressor = MLPRegressor(hidden_layer_sizes=(100, 50),  # Specify the hidden layers and neurons
                            activation='relu',  # Activation function for hidden layers
                            solver='adam',  # Optimization solver
                            alpha=0.0001,  # L2 regularization parameter
                            batch_size='auto',  # Mini-batch size
                            learning_rate='constant',  # Learning rate schedule
                            learning_rate_init=0.001,  # Initial learning rate
                            max_iter=1000,  # Maximum number of iterations
                            random_state=22)  # Random seed for reproducibility
```

```python

# Fit the model to the training data
nn_regressor.fit(X_train_all_merged, y_train)
```

```python

# Make predictions on the test data
y_pred_train = nn_regressor.predict(X_train_all_merged)

# Calculate the R-squared score for training data
nn_train_score = nn_regressor.score(X_train_all_merged, y_train)
print(f"R-squared Score for Training Data: {nn_train_score}")
```

```python



# Evaluate the model's performance
mse = mean_squared_error(y_train, y_pred_train)  
print(f"Mean Squared Error for Training Data: {mse}")


y_pred_val = nn_regressor.predict(X_val_all_merged)

nn_val_score = r2_score(y_val, y_pred_val)  
print(f"R-squared Score for Validation Data: {nn_val_score}")


# Calculate the Mean Squared Error for validation data
mse = mean_squared_error(y_val, y_pred_val)  
print(f"Mean Squared Error for Validation Data: {mse}")
```
